```python
#bar차트 2021년 5대광역시 출생아수와 혼인건수
import matplotlib.pyplot as plt

도시 = ['부산', '대구','인천','광주','대전']
출생아수 = [14446, 10661, 14947, 7956, 7414]
혼인건수 = [11081, 7287, 10881, 4901, 5419]

plt.bar(도시, 출생아수, label='출생아수', alpha=0.5, width=0.5)
#차트는 하나임, 항목두개가. 하나는 바, 하나는 플롯으로 나오는거임.
#알파는 투명도, 데이터가 겹칠때 잘보이라고 플롯에는 알파없음.
plt.plot(도시, 혼인건수, label='혼인건수', linestyle='--', marker='o', color='r')
#바차트는 라인스타일이라는 개념이 없음. 그래서 오류뜸., 플롯은 데이터 하나만 있어도 됨. x축 강제로 넘버링 될뿐. 단 바는 X,Y축이 값이 있어야됨.
plt.legend()

plt.title('5대 광역시 혼인건수와 출생아수')
plt.xlabel('도시')
plt.ylabel('혼인(건수) 및 출생아(수)')
plt.show()
```

```python
import matplotlib.pyplot as plt

도시 = ['부산', '대구','인천','광주','대전']
출생아수 = [14446, 10661, 14947, 7956, 7414]
혼인건수 = [11081, 7287, 10881, 4901, 5419]

# 첫 번째 원 그래프 (출생아수)
plt.subplot(1, 2, 1) # 1행 2열의 첫 번째 위치
plt.pie(출생아수, labels=도시, autopct='%1.1f%%', startangle=140)
plt.title('출생아수')

# 두 번째 원 그래프 (혼인건수)
plt.subplot(1, 2, 2) # 1행 2열의 두 번째 위치
plt.pie(혼인건수, labels=도시, autopct='%1.1f%%', startangle=140)
plt.title('혼인건수')

# 전체 그래프의 제목 설정
plt.suptitle('5대 광역시 혼인건수와 출생아수')

# 그래프 출력
plt.show()
```
