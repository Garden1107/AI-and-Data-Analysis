```python
import pandas as pd
a = pd.read_csv('중간기말성적.csv' , encoding='cp949')
```

```python
#아래 껏들을 잘 알고 있어야함. 모르면 수업에 관심이 없는것.

#a변수에 있는 학번을 빼고 표시, 만 하는 코드
a.drop(columns=['학번'], axis=1)

#a변수에 있는 데이터 중 학번 열을 빼고 표시된 결과를 a변수에 덮어쓴다. inplace=True
a.drop(columns=['학번'], axis=1, inplace=True)

#a변수에 있는 데이터중 학번열을 뺴고 표시된 결과를 b변수에 넣음   #a에는 학번이 안빠져있음.
b=a.drop(columns=['학번'], axis=1)
```
