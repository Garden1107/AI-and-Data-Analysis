```python
import folium

# 지도를 생성
map = folium.Map(location=[37.4955, 126.9560])

# 마커를 생성하고 지도에 추가
folium.Marker(location=[37.4955, 126.9560]).add_to(map)

# 지도를 표시
map.save("my_map.html")
```

```python
folium.Marker(location=[37.4955,126.9560], tooltip='형남공학',).add_to(map)    #마커를 가로로 표시

folium.Marker(location=[37.4955,126.9560], popuo='고민사거리',).add_to(map)     #마커를 세로로 표시

mk2 = folium.Marker(location = [37.50,126.95], tooltip='고민사거리')
mk2.add_to(map)

mk3 = folium.Marker([37.51,126.94], popup='위치3', icon=folium.Icon())
mk3.add_to(map)

mk4 = folium.Marker([37.52,126.96], popup='위치4', icon=folium.Icon(color='red', icon='hippo', prefix='fa'))
mk4.add_to(map)

#원으로 marker 표시하기
mk5 = folium.Circle([37.53,126.97], popup='위치5', color='green')
mk5.add_to(map)

mk6 = folium.Circle([37.54,126.98], popup='위치6', color='green', radius=100)
mk6.add_to(map)

mk7 = folium.Circle([37.51,126.95], popup='위치7', color='green', radius=100, fill=True)
mk7.add_to(map)
```
